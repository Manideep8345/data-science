```python
import cv2
import numpy as np
nam="IMGi.jpg"
img=cv2.imread(nam)#,cv2.IMREAD_REDUCED_COLOR_2)
imj=cv2.imread(nam)#,cv2.IMREAD_REDUCED_COLOR_2)
imyu=cv2.imread(nam)#,cv2.IMREAD_REDUCED_COLOR_2)
lap=cv2.Canny(img,100,150)
#cv2.rectangle(img,(15,25),(200,150),(0,0,0),1)
#lap =cv2.imread(res)
iar=np.asarray(lap)
r=0;x=0;y=0
for eachrow in iar:
    x=x+1;y=0
    for pi in eachrow:
        y=y+1
        
        if(pi>250):
            cv2.line(img,(y,x),(y,x),(25,228,5),1)
            cv2.line(imyu,(y,x),(y,x),(0,0,0),5)
            r=r+1
        else:
            cv2.line(imyu,(y,x),(y,x),(47,228,245),1)
print "no. of pixles "+str(r)
print "height: "+str(x)
print "width: "+str(y)
cv2.imshow('orgl',imj)
cv2.imshow('edge_drawn',img)
cv2.imshow('edge',lap)
cv2.imshow('edge_ty',imyu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imi=cv2.imread('lki.jpg',cv2.IMREAD_REDUCED_COLOR_4)
blur=cv2.blur(imi,(5,5))
bi_blur=cv2.bilateralFilter(imi,9,175,175)#removes noise while preserving edges
m_blur=cv2.medianBlur(imi,9)
bi_edge=cv2.Canny(bi_blur,69,69)
edge=cv2.Canny(imi,250,250)
gui_blur=cv2.GaussianBlur(imi,(11,11),0) #size of kernal should be +ve and odd
cv2.imshow('orgl',imi)
cv2.imshow('bi_edge',bi_edge)
cv2.imshow('edge',edge)
cv2.imshow('bi_blur',bi_blur)
cv2.imshow('m_blur',m_blur)
cv2.imshow('g_blur',gui_blur)
cv2.imshow('blur',blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
